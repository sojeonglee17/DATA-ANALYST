```python
#1사분위수 / 3사분위수

Q1=mpg['cty'].quantile(0.25)#14
Q3=mpg['cty'].quantile(0.75)#19

IQR=Q3-Q1

#상한 / 하한

pct_Q1=Q1-(IQR*1.5)#6.5
pct_Q3=Q3+(IQR*1.5)#26.5

#상한 하한을 벗어난 부분(이상치) 결측처리
mpg['cty']=np.where((mpg['cty']<6.5) | (mpg['cty']>26.5),np.nan,mpg['cty'])

#다시 상자 그림 확인
sns.boxplot(data=mpg,y='cty')
```

```python
#개인 저축률 psavert 시계열 그래프

economics=pd.read_csv('economics.csv')


#date를 날짜 시간 타입 변수 만들기 
economics['date2']=pd.to_datetime(economics['date'])
#연도 변수 만들기
economics['year']=economics['date2'].dt.year
#x축 연도 표시 시계열 그래프 만들기
sns.lineplot(data=economics,x='year',y='psavert',errorbar=None)

```

```python
#2014년 월별 psanert변화 시계열 그래프

economics['month']=economics['date2'].dt.month
date_2014 =economics.query('year==2014')#dataset

sns.lineplot(data=date_2014,x='month' ,y='psavert')
```
